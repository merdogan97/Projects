```python

def bar_plot(variable):
    """
        input : variable ex:"category"
        output: bar plot & value count
    """
    var = df[variable]         
    varValue= var.value_counts()     
   
    plt.figure(figsize=(16,6))
    plt.ticklabel_format(style= "plain", axis="y") 
    plt.xticks(rotation=45);
    plt.bar(varValue.index, varValue)
    plt.xticks(varValue.index, varValue.index.values)
    plt.ylabel("Frequency")
    plt.title(variable)
   
    plt.show()
    print("{}: \n {}".format(variable,varValue))
```

```python
category1 = ["category","installs"]
for c in category1:
    bar_plot(c)
```

```python
def plot_hist(variable):
    
    plt.figure(figsize=(9,3)) 
    plt.hist(df[variable], bins = 100)   
    plt.xlabel(variable)
    plt.ylabel("Frequency")
    plt.title("{} distribution with hist".format(variable))
    plt.show()
```

```python
numericVar = ["1 star ratings", "2 star ratings", "3 star ratings" ,"4 star ratings", "5 star ratings","total ratings","average rating","growth (30 days)","growth (60 days)"]

for n in numericVar: 
    plot_hist(n)
```
